```python
import pandas as pd
```

```python
lego_sets = pd.read_csv('datasets/lego_sets.csv')
parent_themes = pd.read_csv('datasets/parent_themes.csv')

lego_sets = lego_sets.merge(parent_themes, left_on='parent_theme', right_on='name')
```

```python
lego_sets.dtypes
```

```python
# filter for licensed sets
licensed = lego_sets[lego_sets['is_licensed']=True]]
#lego_sets.dropna(subset=['set_num'], inplace=True)
```

**1. What percentage of all _licensed_ sets ever released were Star Wars themed?**  

```python
# drop missing values and filter by theme
sw_sets = lego_sets.query('parent_theme == "Star Wars"')
```

```python
sw_sets.head()
```

```python
# check for duplicates
dupes = sw_sets.duplicated()
len(dupes[dupes == True])
```

```python
# percentage of licenced sets that are star wars
the_force = round((sw_sets.shape[0]/lego_sets.shape[0])*100)
assert isinstance(the_force, int)
```

```python
print(f"""The percentage of all licensed lego sets ever \
released that were star wars themed:\n{the_force}%""")
```
